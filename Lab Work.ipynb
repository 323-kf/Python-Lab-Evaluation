### Q5. List all subjects which has highest and lowest passing and failure rate, respectively.

```python
dict = {}
for i in df.columns[2:10]:
    pass_count = len([j for j  in df[i] if(j >= 3.0)])
    fail_count = len([j for j in df[i] if(j < 3.0)])
    dict[i] = [pass_count, fail_count, np.round((pass_count/df[i].count())*100,2), np.round((fail_count/df[i].count())*100,2)]
pd.DataFrame(dict,index = ['Passed_Count', 'Failed_count', 'Passing Rate', 'Failing Rate'])
```

#### As we can observe from the above table that:

#### Civic Education has highest passing rate and History has highest failure rate.

### Q11. Calculate mean, median, mode, standard deviation and variance of the dataset, and for each subject.
